```python
# ============================================================================
# WavLM Ultimate Architecture
# 核心改进: 
# 1. Head: Layer-wise Weighted Sum -> Bi-LSTM -> Attention Pooling
# 2. Loss: Focal Loss (解决难样本挖掘)
# 3. 策略: 分层学习率 + 冻结策略
# ============================================================================

import os
# 离线模式配置
os.environ["HF_ENDPOINT"] = "https://hf-mirror.com"
os.environ["TRANSFORMERS_OFFLINE"] = "1" 

import numpy as np
import pandas as pd
import librosa
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from pathlib import Path
from sklearn.model_selection import GroupShuffleSplit
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix, f1_score
from transformers import Wav2Vec2FeatureExtractor, WavLMModel
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

# 检查 GPU
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# =====================
# 1. 配置参数
# =====================
MODEL_NAME_OR_PATH = "../wavlm_base_plus_local"
AUDIO_DIR = Path("../AudioWAV")
OUTPUT_DIR = "../Output/WavLM_Ultimate_LSTM_Attn"
os.makedirs(OUTPUT_DIR, exist_ok=True)

CONFIG = {
    "sr": 16000,
    "max_duration": 3.0,
    "batch_size": 128,     
    "num_epochs": 50,     # 结构变复杂了，多给点时间收敛
    "lr": 1e-4,           # Head 学习率
    "backbone_lr": 1e-5,  # Backbone 学习率
    "freeze_layers": 8,   # 冻结前 8 层
    "lstm_hidden": 256,   # LSTM 隐藏层维度
    "num_classes": 6
}

# =====================
# 2. 数据准备 (保持不变)
# =====================
def build_metadata(audio_dir):
    records = []
    for wav in audio_dir.glob("*.wav"):
        parts = wav.stem.split("_")
        if len(parts) == 4:
            records.append({"path": str(wav), "spk": parts[0], "emo": parts[2]})
    return pd.DataFrame(records)

meta = build_metadata(AUDIO_DIR)
le = LabelEncoder()
meta['label'] = le.fit_transform(meta['emo'])
gss = GroupShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
tr_idx, te_idx = next(gss.split(meta, meta['label'], groups=meta['spk']))
train_df = meta.iloc[tr_idx].reset_index(drop=True)
test_df = meta.iloc[te_idx].reset_index(drop=True)
print(f"Train: {len(train_df)}, Test: {len(test_df)}")

# =====================
# 3. Dataset (保持不变)
# =====================
class CREMADataset(Dataset):
    def __init__(self, df, processor, sr=16000, duration=3.0, augment=False):
        self.df = df; self.processor = processor; self.sr = sr
        self.target_len = int(sr * duration); self.augment = augment

    def __len__(self): return len(self.df)

    def __getitem__(self, idx):
        path = self.df.iloc[idx]['path']
        wav, _ = librosa.load(path, sr=self.sr)
        
        # Padding/Cropping
        if len(wav) > self.target_len:
            start = np.random.randint(0, len(wav) - self.target_len) if self.augment else (len(wav) - self.target_len) // 2
            wav = wav[start : start + self.target_len]
        else:
            wav = np.pad(wav, (0, self.target_len - len(wav)), mode='constant')

        # Simple Augmentation
        if self.augment and np.random.random() < 0.3:
            wav += 0.005 * np.random.randn(len(wav))

        inputs = self.processor(wav, sampling_rate=self.sr, return_tensors="pt")
        return {"input_values": inputs.input_values.squeeze(0), "labels": torch.tensor(self.df.iloc[idx]['label'], dtype=torch.long)}

def collate_fn(batch):
    return {
        "input_values": torch.stack([item['input_values'] for item in batch]),
        "labels": torch.stack([item['labels'] for item in batch])
    }

processor = Wav2Vec2FeatureExtractor.from_pretrained(MODEL_NAME_OR_PATH, local_files_only=True)
train_loader = DataLoader(CREMADataset(train_df, processor, augment=True), batch_size=CONFIG['batch_size'], shuffle=True, collate_fn=collate_fn, num_workers=4)
test_loader = DataLoader(CREMADataset(test_df, processor, augment=False), batch_size=CONFIG['batch_size'], shuffle=False, collate_fn=collate_fn, num_workers=4)

# =====================
# 4. 关键组件: Attention Pooling & Focal Loss
# =====================

class SelfAttentionPooling(nn.Module):
    """
    自动学习每一帧的权重
    Input: (Batch, Time, Dim)
    Output: (Batch, Dim)
    """
    def __init__(self, input_dim):
        super().__init__()
        self.W = nn.Linear(input_dim, 1)
        self.softmax = nn.Softmax(dim=1)
        
    def forward(self, batch_rep):
        # (B, T, 1)
        att_logits = self.W(batch_rep)
        # (B, T, 1)
        att_weights = self.softmax(att_logits)
        # Weighted Sum -> (B, D)
        utt_rep = torch.sum(batch_rep * att_weights, dim=1)
        return utt_rep

class FocalLoss(nn.Module):
    """
    Focal Loss: 降低简单样本权重，关注难样本
    Loss = -alpha * (1 - p)^gamma * log(p)
    """
    def __init__(self, gamma=2.0, alpha=None, reduction='mean'):
        super().__init__()
        self.gamma = gamma
        self.alpha = alpha # alpha 可以是类别权重向量
        self.reduction = reduction

    def forward(self, inputs, targets):
        ce_loss = F.cross_entropy(inputs, targets, reduction='none', weight=self.alpha)
        pt = torch.exp(-ce_loss)
        focal_loss = (1 - pt) ** self.gamma * ce_loss
        
        if self.reduction == 'mean':
            return focal_loss.mean()
        elif self.reduction == 'sum':
            return focal_loss.sum()
        else:
            return focal_loss

# =====================
# 5. 模型架构: WavLM + Bi-LSTM + Attn
# =====================
class WavLMUltimateModel(nn.Module):
    def __init__(self, model_name, num_classes, freeze_layers=8, lstm_hidden=256):
        super().__init__()
        
        # 1. Backbone
        try:
            self.wavlm = WavLMModel.from_pretrained(model_name, local_files_only=True, output_hidden_states=True)
        except:
            self.wavlm = WavLMModel.from_pretrained("microsoft/wavlm-base-plus", output_hidden_states=True)
            
        # Freeze
        self.wavlm.feature_extractor._freeze_parameters()
        for layer in self.wavlm.encoder.layers[:freeze_layers]:
            for param in layer.parameters():
                param.requires_grad = False
        
        # Layer Weights
        self.layer_weights = nn.Parameter(torch.ones(13) / 13) 
        
        # 2. Bi-LSTM Head
        self.lstm = nn.LSTM(
            input_size=768, 
            hidden_size=lstm_hidden, 
            num_layers=2, 
            batch_first=True, 
            bidirectional=True,
            dropout=0.2 # LSTM 内部 Dropout
        )
        
        # 3. Attention Pooling
        self.attn_pooling = SelfAttentionPooling(lstm_hidden * 2)
        
        # 4. Classifier
        self.classifier = nn.Sequential(
            nn.Linear(lstm_hidden * 2, 256),
            nn.BatchNorm1d(256), # 加个 BN 稳一点
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(256, num_classes)
        )

    def forward(self, input_values):
        outputs = self.wavlm(input_values)
        
        # Layer Weighted Sum
        all_layers = torch.stack(outputs.hidden_states, dim=-1) # (B, T, 768, 13)
        norm_weights = F.softmax(self.layer_weights, dim=0)
        weighted_feature = (all_layers * norm_weights).sum(dim=-1) # (B, T, 768)
        
        # Bi-LSTM
        lstm_out, _ = self.lstm(weighted_feature) # (B, T, 2*Hidden)
        
        # Attention Pooling
        pooled_feature = self.attn_pooling(lstm_out) # (B, 2*Hidden)
        
        # Classifier
        logits = self.classifier(pooled_feature)
        
        return logits

# =====================
# 6. 训练准备
# =====================
model = WavLMUltimateModel(MODEL_NAME_OR_PATH, CONFIG['num_classes'], 
                          freeze_layers=CONFIG['freeze_layers'], 
                          lstm_hidden=CONFIG['lstm_hidden'])
model.to(device)

# 计算类别权重 (可选，如果 Focal Loss 效果不好再加 alpha)
# class_counts = meta['label'].value_counts().sort_index().values
# class_weights = 1.0 / class_counts
# class_weights = torch.FloatTensor(class_weights / class_weights.sum()).to(device)

# 使用 Focal Loss
criterion = FocalLoss(gamma=2.0) # alpha=class_weights (可选)

# 优化器
optimizer = torch.optim.AdamW([
    {'params': model.wavlm.parameters(), 'lr': CONFIG['backbone_lr']},
    {'params': model.layer_weights, 'lr': 1e-3},
    {'params': model.lstm.parameters(), 'lr': CONFIG['lr']},
    {'params': model.attn_pooling.parameters(), 'lr': CONFIG['lr']},
    {'params': model.classifier.parameters(), 'lr': CONFIG['lr']}
], weight_decay=1e-3) # 稍微加大 Weight Decay

# 记录
best_f1 = 0.0
save_path = os.path.join(OUTPUT_DIR, "best_wavlm_lstm.pth")

print("\nStarting Training...")

# =====================
# 7. 训练循环
# =====================
from tqdm import tqdm

for epoch in range(CONFIG['num_epochs']):
    model.train()
    train_loss = 0
    all_preds, all_labels = [], []
    
    pbar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{CONFIG['num_epochs']}", leave=False)
    
    for batch in pbar:
        inputs = batch['input_values'].to(device)
        labels = batch['labels'].to(device)
        
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        
        # 梯度裁剪 (防止 LSTM 梯度爆炸)
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
        
        optimizer.step()
        
        train_loss += loss.item()
        _, predicted = torch.max(outputs.data, 1)
        all_preds.extend(predicted.cpu().numpy())
        all_labels.extend(labels.cpu().numpy())
        
        pbar.set_postfix({"loss": f"{loss.item():.4f}"})
    
    train_acc = np.mean(np.array(all_preds) == np.array(all_labels))
    
    # --- Validation ---
    model.eval()
    val_preds = []
    val_labels = []
    
    with torch.no_grad():
        for batch in test_loader:
            inputs = batch['input_values'].to(device)
            labels = batch['labels'].to(device)
            outputs = model(inputs)
            _, predicted = torch.max(outputs.data, 1)
            val_preds.extend(predicted.cpu().numpy())
            val_labels.extend(labels.cpu().numpy())
            
    val_f1 = f1_score(val_labels, val_preds, average='macro')
    val_acc = np.mean(np.array(val_preds) == np.array(val_labels))
    
    print(f"Epoch {epoch+1} | Loss: {train_loss/len(train_loader):.4f} | "
          f"Tr Acc: {train_acc*100:.2f}% | Val Acc: {val_acc*100:.2f}% | F1: {val_f1:.4f}")
    
    # Save
    if val_f1 > best_f1:
        best_f1 = val_f1
        torch.save(model.state_dict(), save_path)
        print(f"  ★ New Best F1! Model saved.")

# =====================
# 8. 最终评估 & 绘图
# =====================
print("\nLoading best model...")
model.load_state_dict(torch.load(save_path))
model.eval()

y_pred, y_true = [], []
with torch.no_grad():
    for batch in test_loader:
        inputs = batch['input_values'].to(device)
        labels = batch['labels'].to(device)
        outputs = model(inputs)
        _, predicted = torch.max(outputs.data, 1)
        y_pred.extend(predicted.cpu().numpy())
        y_true.extend(labels.cpu().numpy())

print("\nClassification Report:")
print(classification_report(y_true, y_pred, target_names=le.classes_, digits=4))

# 混淆矩阵
cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=le.classes_, yticklabels=le.classes_)
plt.title("Confusion Matrix (WavLM + LSTM + Attn)")
plt.savefig(os.path.join(OUTPUT_DIR, "confusion_matrix.png"))
plt.show()

# 权重热力图
weights = F.softmax(model.layer_weights, dim=0).detach().cpu().numpy()
plt.figure(figsize=(10, 4))
sns.barplot(x=list(range(13)), y=weights, palette="viridis")
plt.title("Learned Layer Weights")
plt.savefig(os.path.join(OUTPUT_DIR, "layer_weights.png"))
```
