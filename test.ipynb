```python
import tensorflow as tf

gpus = tf.config.list_physical_devices('GPU')
print("物理 GPU:", gpus)

for i, gpu in enumerate(gpus):
    try:
        logical_gpus = tf.config.list_logical_devices('GPU')
        print(f"逻辑 GPU {i}:", logical_gpus)
    except Exception as e:
        print("无法获取逻辑 GPU：", e)
```
