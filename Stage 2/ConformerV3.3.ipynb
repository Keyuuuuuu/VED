```python
# ============================================================================
# Conformer V3.3 - The "Trident" Architecture & SAM Optimizer
# 改进点:
# 1. Input: 3 Channels (Log-Mel + Delta + Delta-Delta)
# 2. Regularization: Stochastic Depth (DropPath)
# 3. Optimizer: SAM (Sharpness-Aware Minimization)
# ============================================================================

import os
import math
import numpy as np
import pandas as pd
import librosa
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
from datetime import datetime
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import GroupShuffleSplit
from sklearn.metrics import classification_report, confusion_matrix, f1_score
import tensorflow as tf
from tensorflow.keras import layers, models, callbacks

# 随机种子
SEED = 42
np.random.seed(SEED)
tf.random.set_seed(SEED)

# 1. 目录与配置
def create_experiment_dir(base_dir="Output"):
    try: script_name = os.path.splitext(os.path.basename(__file__))[0]
    except: script_name = "conformer_v3_3_notebook"
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    exp_dir = os.path.join(base_dir, f"{script_name}_{timestamp}")
    for d in ["models", "plots", "logs"]: os.makedirs(os.path.join(exp_dir, d), exist_ok=True)
    return exp_dir, os.path.join(exp_dir, "models"), os.path.join(exp_dir, "plots"), os.path.join(exp_dir, "logs")

EXP_DIR, MODEL_DIR, PLOT_DIR, LOG_DIR = create_experiment_dir()
AUDIO_DIR = Path("../AudioWAV")

# 2. 特征提取 (升级为 3 通道)
SR = 16000; N_MELS = 80; FFT = 1024; HOP = 160; WIN = 400; FIXED_SECONDS = 3.0
MAX_FRAMES = int(math.ceil(FIXED_SECONDS * SR / HOP))

def load_feature_3ch(path):
    y, _ = librosa.load(path, sr=SR, mono=True)
    tgt = int(FIXED_SECONDS * SR)
    if len(y) < tgt: y = np.pad(y, (0, tgt - len(y)))
    else: y = y[:tgt]
    
    # 1. Static Log-Mel
    S = librosa.feature.melspectrogram(y=y, sr=SR, n_fft=FFT, hop_length=HOP, n_mels=N_MELS)
    log_S = librosa.power_to_db(S, ref=np.max)
    
    # 2. Delta
    delta = librosa.feature.delta(log_S)
    
    # 3. Delta-Delta
    delta2 = librosa.feature.delta(log_S, order=2)
    
    # Stack -> (Time, Freq, 3)
    # Librosa output is (Freq, Time), so we transpose
    feat = np.stack([log_S.T, delta.T, delta2.T], axis=-1) 
    
    if feat.shape[0] < MAX_FRAMES:
        feat = np.pad(feat, ((0, MAX_FRAMES-feat.shape[0]), (0,0), (0,0)))
    else: feat = feat[:MAX_FRAMES]
    
    return feat.astype(np.float32)

# 3. 加载数据
def build_metadata(audio_dir):
    records = []
    for wav in audio_dir.glob("*.wav"):
        p = wav.stem.split("_")
        if len(p)==4: records.append({"path":wav, "spk":p[0], "emo":p[2]})
    return pd.DataFrame(records)

meta = build_metadata(AUDIO_DIR)
print(f"Extracting 3-Channel Features for {len(meta)} samples...")
X = np.stack([load_feature_3ch(row['path']) for _, row in meta.iterrows()])
print(f"X shape: {X.shape}") # (N, 300, 80, 3)

# 4. 划分与归一化
le = LabelEncoder()
y = le.fit_transform(meta['emo'])
groups = meta['spk'].values
n_classes = len(le.classes_)

gss = GroupShuffleSplit(n_splits=1, test_size=0.2, random_state=SEED)
tr_idx, te_idx = next(gss.split(X, y, groups=groups))
Xtr, Xte = X[tr_idx], X[te_idx]
ytr, yte = y[tr_idx], y[te_idx]

# 逐通道归一化
mean_val = Xtr.mean(axis=(0,1,2), keepdims=True)
std_val = Xtr.std(axis=(0,1,2), keepdims=True) + 1e-6
Xtr = (Xtr - mean_val) / std_val
Xte = (Xte - mean_val) / std_val

# =====================
# 5. 核心组件升级
# =====================

# Stochastic Depth Layer
class DropPath(layers.Layer):
    def __init__(self, drop_prob=0.0, **kwargs):
        super().__init__(**kwargs)
        self.drop_prob = drop_prob
    def call(self, x, training=False):
        if self.drop_prob == 0. or not training: return x
        keep_prob = 1 - self.drop_prob
        shape = (tf.shape(x)[0],) + (1,) * (len(tf.shape(x)) - 1)
        random_tensor = keep_prob + tf.random.uniform(shape, dtype=x.dtype)
        random_tensor = tf.floor(random_tensor)
        return (x / keep_prob) * random_tensor

# SAM Model Wrapper (最关键的部分)
class SAMModel(tf.keras.Model):
    def __init__(self, inputs, outputs, rho=0.05):
        super().__init__(inputs, outputs)
        self.rho = rho # 邻域大小

    def train_step(self, data):
        x, y = data
        # 1. 第一步：计算梯度
        with tf.GradientTape() as tape:
            y_pred = self(x, training=True)
            loss = self.compiled_loss(y, y_pred, regularization_losses=self.losses)
        
        # 计算 epsilon (扰动)
        trainable_vars = self.trainable_variables
        gradients = tape.gradient(loss, trainable_vars)
        norm = tf.linalg.global_norm(gradients)
        scale = self.rho / (norm + 1e-12)
        
        # 2. 施加扰动 (e_w)
        e_ws = []
        for v, g in zip(trainable_vars, gradients):
            e_w = g * scale
            v.assign_add(e_w)
            e_ws.append(e_w)
            
        # 3. 第二步：在扰动后的权重上计算梯度
        with tf.GradientTape() as tape:
            y_pred = self(x, training=True)
            loss_sam = self.compiled_loss(y, y_pred, regularization_losses=self.losses)
            
        sam_gradients = tape.gradient(loss_sam, trainable_vars)
        
        # 4. 撤销扰动并更新权重
        for v, e_w in zip(trainable_vars, e_ws):
            v.assign_sub(e_w)
            
        self.optimizer.apply_gradients(zip(sam_gradients, trainable_vars))
        
        # 更新指标
        self.compiled_metrics.update_state(y, y_pred)
        return {m.name: m.result() for m in self.metrics}

# 其他组件 (保持 V3.2 优秀部分)
class SpecAugment(layers.Layer):
    def call(self, x, training=False):
        if not training: return x
        B, T, F, C = tf.shape(x)[0], tf.shape(x)[1], tf.shape(x)[2], tf.shape(x)[3]
        # 只 Mask 前两个维度，保留 Channel 信息
        t = tf.random.uniform([], 0, 20, dtype=tf.int32)
        t0 = tf.random.uniform([], 0, tf.maximum(1, T-t), dtype=tf.int32)
        m1 = tf.concat([tf.ones([B,t0,F,C]), tf.zeros([B,t,F,C]), tf.ones([B,T-t0-t,F,C])], 1)
        f = tf.random.uniform([], 0, 10, dtype=tf.int32)
        f0 = tf.random.uniform([], 0, tf.maximum(1, F-f), dtype=tf.int32)
        m2 = tf.concat([tf.ones([B,T,f0,C]), tf.zeros([B,T,f,C]), tf.ones([B,T,F-f0-f,C])], 2)
        return x * m1 * m2

def glu(x): return x[..., :x.shape[-1]//2] * tf.sigmoid(x[..., x.shape[-1]//2:])
def se_block(x, ratio=8):
    ch = x.shape[-1]
    s = layers.GlobalAveragePooling2D()(x)
    s = layers.Dense(ch//ratio, activation='relu')(s)
    s = layers.Dense(ch, activation='sigmoid')(s)
    return layers.Multiply()([x, layers.Reshape((1,1,ch))(s)])

def multi_scale_conv_block(x):
    reg = tf.keras.regularizers.l2(5e-5)
    c3 = layers.Conv2D(32, 3, padding='same', activation='relu', kernel_regularizer=reg)(x)
    c5 = layers.Conv2D(32, 5, padding='same', activation='relu', kernel_regularizer=reg)(x)
    c7 = layers.Conv2D(32, 7, padding='same', activation='relu', kernel_regularizer=reg)(x)
    x = layers.Concatenate()([c3, c5, c7])
    x = layers.BatchNormalization()(x)
    x = se_block(x)
    return layers.MaxPooling2D((2,2))(x)

def conformer_block(x, d_model=128, heads=4, kernel_size=31, dropout=0.1, drop_path=0.1):
    # FFN 1
    r = x
    x = layers.Dense(d_model*4, activation='swish')(x)
    x = layers.Dropout(dropout)(x)
    x = layers.Dense(d_model)(x)
    x = DropPath(drop_path)(x) # Stochastic Depth
    x = layers.LayerNormalization()(r + 0.5*x)
    
    # MHSA
    r = x
    x = layers.MultiHeadAttention(heads, d_model//heads, dropout=dropout)(x, x)
    x = DropPath(drop_path)(x)
    x = layers.LayerNormalization()(r + x)
    
    # Conv (Large Kernel)
    r = x
    x = layers.Conv1D(d_model*2, 1)(x)
    x = layers.Lambda(glu)(x)
    x = layers.DepthwiseConv1D(kernel_size, padding='same')(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('swish')(x)
    x = layers.Conv1D(d_model, 1)(x)
    x = layers.Dropout(dropout)(x)
    x = DropPath(drop_path)(x)
    x = layers.LayerNormalization()(r + x)
    
    # FFN 2
    r = x
    x = layers.Dense(d_model*4, activation='swish')(x)
    x = layers.Dropout(dropout)(x)
    x = layers.Dense(d_model)(x)
    x = DropPath(drop_path)(x)
    x = layers.LayerNormalization()(r + 0.5*x)
    return x

# =====================
# 6. 构建 V3.3 模型
# =====================
def build_conformer_v3_3(input_shape, n_classes):
    inp = layers.Input(input_shape) # (300, 80, 3)
    x = layers.GaussianNoise(0.02)(inp)
    x = SpecAugment()(x)
    
    # CNN (3 channels input handled automatically)
    x = multi_scale_conv_block(x)
    
    # Downsampling
    reg = tf.keras.regularizers.l2(5e-5)
    
    # Block 1
    x = layers.Conv2D(128, 3, padding='same', activation='relu', kernel_regularizer=reg)(x)
    x = layers.BatchNormalization()(x) 
    x = se_block(x)
    x = layers.MaxPooling2D((2,2))(x)
    x = layers.SpatialDropout2D(0.2)(x)
    
    # Block 2
    x = layers.Conv2D(256, 3, padding='same', activation='relu', kernel_regularizer=reg)(x)
    x = layers.BatchNormalization()(x)
    x = se_block(x)
    x = layers.MaxPooling2D((2,2))(x)
    x = layers.SpatialDropout2D(0.2)(x)
    
    # Reshape
    x = layers.Reshape((-1, x.shape[-1]))(x)
    x = layers.Dense(144)(x)
    
    # Conformer with DropPath
    x = conformer_block(x, d_model=144, kernel_size=15, drop_path=0.05)
    x = conformer_block(x, d_model=144, kernel_size=31, drop_path=0.1) 
    
    # Attentive Pooling
    att = layers.Dense(1, activation='tanh')(x)
    att = layers.Softmax(axis=1)(att)
    x = tf.reduce_sum(x * att, axis=1)
    
    x = layers.Dropout(0.3)(x)
    x = layers.Dense(64, activation='relu', kernel_regularizer=reg)(x)
    x = layers.Dropout(0.2)(x)
    out = layers.Dense(n_classes, activation='softmax', name='emotion')(x)
    
    # Return SAMModel wrapper
    return SAMModel(inputs=inp, outputs=out, rho=0.05)

# =====================
# 7. 训练流程
# =====================
class LabelSmoothingFocalLoss(tf.keras.losses.Loss):
    def __init__(self, smoothing=0.05, gamma=2.0, **kwargs):
        super().__init__(**kwargs); self.smoothing=smoothing; self.gamma=gamma
    def call(self, y_true, y_pred):
        n = tf.cast(tf.shape(y_pred)[-1], tf.float32)
        y_true = y_true * (1-self.smoothing) + self.smoothing/n
        y_pred = tf.clip_by_value(y_pred, 1e-7, 1-1e-7)
        return tf.reduce_sum(-y_true*tf.math.log(y_pred)*tf.pow(1-y_pred, self.gamma), axis=-1)

# Mixup Gen
class MixupGen(tf.keras.utils.Sequence):
    def __init__(self, x, y, batch=64):
        self.x, self.y, self.batch = x, y, batch
        self.idx = np.arange(len(x)); np.random.shuffle(self.idx)
    def __len__(self): return int(np.ceil(len(self.x)/self.batch))
    def __getitem__(self, i):
        inds = self.idx[i*self.batch:(i+1)*self.batch]
        bx, by = self.x[inds].copy(), self.y[inds].copy()
        if np.random.random()<0.5:
            lam = np.random.beta(0.2, 0.2)
            p = np.random.permutation(len(bx))
            bx = lam*bx + (1-lam)*bx[p]; by = lam*by + (1-lam)*by[p]
        return bx, by
    def on_epoch_end(self): np.random.shuffle(self.idx)

# Cosine Decay
def lr_schedule(epoch):
    lr_max, lr_min = 8e-4, 1e-6
    if epoch < 5: return lr_max * (epoch+1)/5
    else: return lr_min + 0.5*(lr_max-lr_min)*(1+np.cos(np.pi*(epoch-5)/95))

model = build_conformer_v3_3(Xtr.shape[1:], n_classes)
# 这里的 optimizer 将被 SAM 使用
model.compile(
    optimizer=tf.keras.optimizers.AdamW(learning_rate=8e-4, weight_decay=2e-4),
    loss=LabelSmoothingFocalLoss(smoothing=0.05),
    metrics=['accuracy']
)

# Callbacks
class F1Cb(callbacks.Callback):
    def on_epoch_end(self, e, logs):
        p = np.argmax(self.model.predict(Xte, verbose=0), axis=1)
        f1 = f1_score(yte, p, average='macro')
        print(f" — val_f1: {f1:.4f}")
        logs['val_macro_f1'] = f1

callbacks_list = [
    F1Cb(),
    callbacks.ModelCheckpoint(os.path.join(MODEL_DIR, "best.h5"), monitor='val_macro_f1', mode='max', save_best_only=True, verbose=1),
    callbacks.EarlyStopping(monitor='val_macro_f1', mode='max', patience=30, restore_best_weights=True),
    callbacks.LearningRateScheduler(lr_schedule),
    callbacks.CSVLogger(os.path.join(LOG_DIR, "log.csv"))
]

print("Starting V3.3 Training (SAM + 3-Channel + DropPath)...")
model.fit(
    MixupGen(Xtr, tf.keras.utils.to_categorical(ytr), 64),
    validation_data=(Xte, tf.keras.utils.to_categorical(yte)),
    epochs=100,
    callbacks=callbacks_list,
    verbose=1
)

# =====================
# 8. 评估
# =====================
print("\nEvaluating V3.3...")
p = np.argmax(model.predict(Xte, verbose=0), axis=1)
print(classification_report(yte, p, target_names=le.classes_, digits=4))

cm = confusion_matrix(yte, p)
plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=le.classes_, yticklabels=le.classes_)
plt.title("Confusion Matrix (V3.3 SAM)")
plt.savefig(os.path.join(PLOT_DIR, "cm.png"))
plt.show()

print(f"Results saved to {EXP_DIR}")
```
